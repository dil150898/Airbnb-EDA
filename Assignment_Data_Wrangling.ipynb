```python
# This function will Covert string value to List to do operations
def convert_to_list(string):
  return eval(string)
```

```python
# Converting cast string value to list with the help of above function
imdb_df.loc[~imdb_df['cast'].isna() , 'cast'] = imdb_df.loc[~imdb_df['cast'].isna() , 'cast'].apply(convert_to_list)
```

```python
# check type of cast value which earler coverted
type(imdb_df.loc[0 , 'cast'])
```

```python
# This function will extract the name from the first dictionary of list of each row
def extract_name(cast_list):
    if cast_list and isinstance(cast_list, list):
        first_dict = cast_list[0]
        if 'name' in first_dict:
            return first_dict['name']
    return None
```

```python
# Making a columns Actor name which will store actor name from the help of extract_name function
imdb_df['Actor name'] = imdb_df['cast'].apply(extract_name)
```

```python
# Let's check the null value present in cast data
null_actor_name = imdb_df[imdb_df['Actor name'] . isnull()]
null_actor_name.head()
```

```python
# Let's check which actor have acted in most movies
actor_acted_most_movies = imdb_df['Actor name'].value_counts()
```

```python
# Modifing columns name in actor_acted_most_movies Data Frame
actor_acted_most_movies = imdb_df['Actor name'].value_counts().reset_index()
actor_acted_most_movies.columns = ['Actor name' , 'Nomber of movies']
```

```python
# Let's print who acted in most movies with the help of f string
actor_acted_most_movies_str = (f' The actor acted in most movies is "{actor_acted_most_movies.loc[ 0 , "Actor name"]}". Worked in "{actor_acted_most_movies.loc[ 0 , "Nomber of movies"]}" movies.')
print(actor_acted_most_movies_str)
```

```python
# Let's check the genres data type to do forther operation
type(imdb_df.loc[0 , 'genres'])
```

```python
# Changing genres data type to list with the help of Conver_to_list function which created earlier
imdb_df.loc[~imdb_df['genres'].isna() , 'genres'] = imdb_df.loc[~imdb_df['genres'].isna() , 'genres'].apply(convert_to_list)
```

```python
# Applying extract_name function which created earlier to extract genres type of movies and attaching it into data with new column named genres name
imdb_df['genres name'] = imdb_df['genres'].apply(extract_name)
```

```python
# Let's check movies name
movies_name = imdb_df.loc[imdb_df['Actor name'] == actor_acted_most_movies.loc[0 , 'Actor name'] , 'original_title'].reset_index()
movies_name.drop('index' , axis = 1 , inplace = True)
movies_name
```

```python
# Let's check genres of movies
genres_detail = imdb_df.loc[imdb_df['Actor name'] == actor_acted_most_movies.loc[0 , 'Actor name']][['original_title' , 'genres name']].reset_index()
# Droping extra column
genres_detail.drop('index' , axis = 1 , inplace = True)
genres_detail
```

```python
# Lets check in which types of movies actor acted most
count_types_of_moives_worked_most = genres_detail.groupby("genres name").size().sort_values(ascending = False).reset_index().rename(columns = {0 : 'Nomber of movies'})

# Let's see it with help of graph

# Selecting different different color for each bar
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
# Defining figure size
plt.figure(figsize=(6, 4))
plt.bar(count_types_of_moives_worked_most['genres name'] , count_types_of_moives_worked_most['Nomber of movies'] , color=colors , width=0.6)
plt.xlabel('Types of Movies')
plt.ylabel('Nomber of movies')
# Lets extract name of actor in title the data Correspond to whom
plt.title(f'Nomber of movies worked by {actor_acted_most_movies.loc[ 0 , "Actor name"]}')
plt.xticks(rotation = 45)
plt.show()
```

```python
# Let's check profit of all the movie
imdb_df['profit'] = imdb_df['revenue'] - imdb_df['budget']
movies_detail = imdb_df.loc[imdb_df['Actor name'] == actor_acted_most_movies.loc[0 , 'Actor name']][['original_title' , 'genres name' , 'profit']].reset_index()
movies_detail.drop('index' , axis = 1 , inplace = True)
movies_detail
```

```python
# Above we seen some movies profit is in nigative which means that movies was in loss lets ceck how many movies are in loss
movie_in_loss = movies_detail[movies_detail['profit'] < 0 ].reset_index()
movie_in_loss.drop('index' , axis = 1 , inplace = True)
movie_in_loss
```

```python
# Let's check which movies in each types gain max profit
max_profit_movies = movies_detail.loc[movies_detail.groupby('genres name')['profit'].idxmax()]

# Adding a column & Combining both genres & original title togther for beter understanding in bar graph
max_profit_movies['combined_label'] = max_profit_movies['original_title'] + '\n(' + max_profit_movies['genres name'] + ')'

# Let's plot it in bar
plt.bar(max_profit_movies['combined_label'], max_profit_movies['profit'] , color = colors)
plt.xlabel('Max profit movie name with Genres')
plt.ylabel('Maximum Profit')
plt.title('Maximum Profit in each Genres')
plt.xticks(rotation=45)
plt.show()
```

```python
# Lets see which movie made max profit
max_profit_movie_index = movies_detail['profit'].idxmax()

print(f''''{movies_detail.loc[max_profit_movie_index , 'original_title']}', Which is a '{movies_detail.loc[max_profit_movie_index , 'genres name']}' Movie.
This movie earned '{movies_detail.loc[max_profit_movie_index , 'profit']}', Which is highest profit of '{actor_acted_most_movies.loc[ 0 , "Actor name"]}'.''')
```

```python


def extract_producer_name(crew_list):
    """
    Extracts the names of producers from a list of crew members.

    Args:
        crew_list (list): A list of dictionaries, where each dictionary represents
                          a crew member with 'job' and 'name' keys.

    Returns:
        list: A list of producer names.
    """
    producer_list = []
    # Check if crew_list is iterable (list or tuple) and not a float (NaN)
    if isinstance(crew_list, list):
        for elem in crew_list:
            if isinstance(elem, dict) and elem.get('job') == "Producer":
                producer_list.append(elem.get('name'))
    return producer_list

# Apply the function to the 'crew' column
#imdb_df['Producer name'] = imdb_df['crew'].apply(extract_producer_name)
```
